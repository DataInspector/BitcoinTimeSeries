# Import Packages

```r
library(lubridate)
library(repr)
library(forecast)
library(tidyverse)
library(zoo)
library(xts)
library(fUnitRoots)
library(tseries)
```

# Import Data And Plot Time Series

```r
options(repr.plot.width=4, repr.plot.height=3)
Data <- read.csv("BitcoinTimeSeries.csv")[2:3]
Data$Date <- as.Date(Data$Date)
Data <- Data[order(Data$Date, decreasing = FALSE),]
Data <- as.data.frame(Data)
row.names(Data) <- c()
ggplot(data = Data, mapping = aes(x=Date,y=Close)) + geom_line()
```

# Take The Log & Plot With Moving Average

```r
MAData <- Data
MAData$Close <- log(MAData$Close, base = 5)
MAData$Close <- tsclean(MAData$Close)
MAData$MA <- ma(MAData$Close, order=50)
ggplot() + geom_line(data = MAData, aes(x = Date, y=Close)) + geom_line(data = MAData, aes(x = Date, y=MA), color = "red")
print(min(Data$Date))
```

# Create Time Series & Look At Dfference With Lagged Figures

```r
LagData <- Data
TSLogData <- ts(data = log(LagData$Close, base = 5), start = c(2014,1,1), frequency = 365)
LagData <- cbind(TSLogData, diff(TSLogData, lag = 1))
LagData <- cbind(LagData, diff(TSLogData, lag = 7))
LagData <- cbind(LagData, diff(TSLogData, lag = 30))
plot(LagData)
```

# Augmented Dicky Fuller For Multiple Lags

```r
LagMax <- 365
ADFData <- setNames(data.frame(matrix(ncol = 3, nrow =LagMax)), c("ADF", "PValue","Lag"))
for (i in 1:LagMax){
    ADFData$PValue[i] <- adf.test(TSLogData, alternative = "stationary", k =  i)$p.value #Non Staionary ADF
    ADFData$ADF[i] <- adf.test(TSLogData, alternative = "stationary", k =  i)$statistic #Non Staionary ADF
    ADFData$Lag[i] <- i #Non Staionary ADF
    }
ADFData[ADFData$PValue == min(as.numeric(ADFData$PValue)),]
```

# Auto Correlation Function & Partial AutoCorrelation Function

```r
print(str(TSLogData))
Acf(TSLogData)
Pacf(TSLogData)
```

# Decompose Data To Split Seasonality With Trend & Unexplained

```r
plot(decompose(ts(data = TSLogData, start = c(2014,1,1), frequency = 365)))
decomp = seasadj(stl(ts(data = TSLogData, start = c(2014,1,1), frequency = 365), s.window="periodic"))
plot(decomp)
```

#  Test Time Series Model With Moving Average Data

```r
MAData <- na.omit(MAData)
print(min(MAData$Date))
MATSData <- ts(data = MAData$MA, start = c(2014,2,20), frequency = 365)
plot(MATSData)
```

# Decompose MA Data To Split Seasonality With Trend & Unexplained

```r
TSDecomp <- stl(MATSData, s.window="periodic")
plot(TSDecomp)
decomp = seasadj(TSDecomp)
plot(decomp)
```

# Auto Correlation Function & Partial AutoCorrelation Function

```r
print(str(decomp))
Acf(decomp)
Pacf(decomp)
```

# Differencing & Augmented Dicky Fuller For Multiple Lags

```r
MADiffData = diff(decomp, differences = 1)
plot(MADiffData)
adf.test(MADiffData, alternative = "stationary")
```

# Differencing & Augmented Dicky Fuller For Decomposed Log

```r
DecompDiffData = diff(decomp, differences = 1)
plot(DecompDiffData)
adf.test(DecompDiffData, alternative = "stationary")
```

# Auto Correlation Function & Partial AutoCorrelation Function

```r
print(str(DecompDiffData))
Acf(DecompDiffData)
Pacf(DecompDiffData)
```
